```python
from pprint import pprint
from room_env.envs.room2 import RoomEnv2

env_config = {
    "question_prob": 1.0,
    "seed": 0,
    "terminates_at": 99,
    "randomize_observations": True,
    "room_size": "m",
    "rewards": {"correct": 1, "wrong": -1, "partial": 0},
    "make_everything_static": True,
}

env = RoomEnv2(**env_config)
observations, info = env.reset()
env.render("image", save_fig_dir="./DEBUG/")
for key, val in env.names.items():
    print(f"The number of {key}: {len(val)}")

print()
pprint(observations)
pprint(info)

for obj_type, objs in env.objects.items():
    for obj in objs:
        print(f"{obj_type}\t{obj.name}\t{obj.location}\t{obj.history}".expandtabs(16))
```

```python
for _ in range(10):
    observations, reward, done, truncated, info = env.step(("room_000", "south"))
    env.render("image", save_fig_dir="./DEBUG/")

    print()
    pprint(observations)
    pprint(reward)
    pprint(done)
    pprint
    pprint(info)

    for obj_type, objs in env.objects.items():
        for obj in objs:
            print(
                f"{obj_type}\t{obj.name}\t{obj.location}\t{obj.history}".expandtabs(16)
            )
```
